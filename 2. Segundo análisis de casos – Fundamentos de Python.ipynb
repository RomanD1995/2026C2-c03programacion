# Problema #1: Calculadora de propina inteligente

## Problema
debo calcular el porcentage de propina basado en el monto total de la cuenta 
debo pedir el monto de la cuenta total y cuanto decea dar de propina y calcular la propina y el monto final


## Entrada
- pedir que el ususario ingrese el total de la cuenta en numero decimal 
- pedir el porcentaje de propina 

## Proceso
#manejo de exepciones 
- validamos utilisamos try-except para capturar el erros por sei el ususario ingresa letras
- clacular el monto de propina basado en el monto total 
- pripina = cuenta * (porcentaje / 100)

## Salida esperada
espero generar un mensajse claro con formato de stringgs 
que muestre el monto de la cuenta ,el de propina y el monto total a pagar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |$100            | 10%                | $110               |
| Caso límite |$50.5           | 0%                 | $50.5              |
| Caso inválido |veinte        |15%                 | ingrese numeros    |


```python
 #Ejercicio 1: Calculadora de propina inteligente

print("--- Calculadora de Propina Inteligente ---")

try:
    # Solicitud de datos 
    total_cuenta = float(input("Ingrese el total de la cuenta : "))
    porcentaje_propina = float(input("Ingrese el porcentaje de propina (ej. 10, 15, 20) : "))
    
    # no permito valores negativos 
    if total_cuenta < 0 or porcentaje_propina < 0:
        print("\n[Error]: Los valores no pueden ser negativos.")
    else:
        # Procesamiento de datos
        monto_propina = total_cuenta * (porcentaje_propina / 100)
        total_final = total_cuenta + monto_propina
        
        # Salida de datos 
        print("\n--- Resumen de Pago ---")
        print(f"Subtotal cuenta:  ${total_cuenta:,.2f}")
        print(f"Propina ({porcentaje_propina:.0f}%):     ${monto_propina:,.2f}")
        print(f"Total a pagar:    ${total_final:,.2f}")

except ValueError:
    # Capturamos el error 
    print("\n[Error]: Por favor, ingrese únicamente números válidos (use coma para decimales).")
```

```python
#Ejercicio 2: Detector de número par/impar

print("--- Detector de Número Par o Impar ---")

try:
    # Solicitamos número 
    numero = int(input("Ingrese un número entero: "))
    
    # Proceso de (Par o Impar)
    if numero % 2 == 0:
        print(f"\nEl número {numero} es PAR  exelente . . .")
    else:
        print(f"\nEl número {numero} es IMPAR ya que no es divicible entre 2.")

except ValueError:
    # Captura el error si el usuario ingresa letras o números con punto decimal
    print("\n[Error]: Debe ingresar obligatoriamente un número entero (sin letras ni decimales).")
```

crear un script que te clasifique segun tu edad

## Entrada

## Salida esperada

o un mensaje en caso de error

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |17              | eres menor de edad |menor               |
| Caso límite |18              | eres mayor         |mallor de edad      |
| Caso inválido | -18          | mensajde de error  |ingrese nuemro      |


```python
#Ejercicio 3: Clasificador de edad

print("--- Clasificador de Categoría de Edad ---")

try:
    # Solicitamos la edad 
    edad = int(input("Ingrese su edad en años: "))
    
    # Validación si (no hay edades negativas)
    if edad < 0:
        print("\n[Error]: La edad no puede ser un número negativo o letras .")
    # Clasificación usando (if-elif-else)
    elif edad < 18:
        print(f"\nCon {edad} años, eres un joven : MENOR DE EDAD.")
    elif edad <= 59:
        print(f"\nCon {edad} años,  eres mayor de edad : ADULTO.")
    else:
        print(f"\nCon {edad} años,  señor en este caso eres : ADULTO MAYOR.")

except ValueError:
    # captura el error 
    print("\n[Error]: Por favor, introduzca un número entero válido para tu la edad.")
```

# Problema #4: Validador de compra con descuento

## Problema

debo desaroyar unnscript que valide un descuento a clientes frecuetes ,
y del moto de su compra debe basar un descuento 

## Entrada

- nombre del cliente
- estado del cliente  (nuevo o frecuente )
- monto de la compra
## Proceso
- limpiar los datos con el try exept
- luego evaluamos el monto del cliente para calcular el descuento 
- pero antes vemos si el cliente es frecuente 

## Salida esperada
su monto total 
su deacuento y cual es el monto qeu ahora debe pagar 
ademas de la validacion si es cliente frecuente

- 

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |17.000 y frecuente| decuento del 10%|correcto detecto el decuento  |
| Caso límite |50.000 y  no frecuente|desceunto por monto minimo| si detecto el decuento   |
| Caso inválido |todo correcto pero monto negetivo|detecta el error | envia mensaje de error y como detallle imprime el error|


```python

# Ejercicio 4: Validador de compra con descuento


print("--- Sistema de Validación de Compras ---")

# 1. ENTRADAS: Solicitar datos
nombre_cliente = input(str("Ingrese suu nombre - solo ingrese letras por favor : ")).strip()
es_frecuente_input = input("¿Es cliente frecuente? (sí/no): ").strip().lower()

# Convertir la respuesta 
es_frecuente = es_frecuente_input in ["sí", "si", "s", "yes", "y"]

# 2. PROCESO: Validación del monto 
try:
    monto_compra = float(input("Ingrese el monto : "))
    
    # Validar si es negativo
    if monto_compra < 0:
        raise ValueError("El monto de la compra no puede ser un número negativo.")

    # condiciones 
    porcentaje_descuento = 0.0

    if es_frecuente and monto_compra >= 50000:
        porcentaje_descuento = 0.15 
    elif es_frecuente or monto_compra >= 50000:
        porcentaje_descuento = 0.10 
    else:
        porcentaje_descuento = 0.0  

    # PROCESO: Cálculo de totales 
    descuento_aplicado = monto_compra * porcentaje_descuento
    monto_final = monto_compra - descuento_aplicado

    # 4. SALIDA
    print("\n--- Resumen de la Transacción ---")
    print(f"Cliente: {nombre_cliente}")
    print(f"¿Cliente frecuente?: {'Sí' if es_frecuente else 'No'}")
    print(f"Monto original: ₡{monto_compra:,.2f}")
    print(f"Descuento aplicado ({porcentaje_descuento * 100:.0f}%): ₡{descuento_aplicado:,.2f}")
    print(f"Monto final a pagar: ₡{monto_final:,.2f}")

except ValueError as e:
    # Captura el error 
    print(f"\n[Error]: Entrada inválida. no ingrso un número válido para el monto.")
    print(f"Detalle del error: {e}")
```
