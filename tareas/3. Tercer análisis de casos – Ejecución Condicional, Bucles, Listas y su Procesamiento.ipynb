# Caso 1: Control de ventas diarias de una soda

## Problema
Una soda ubicada en San Pedro registra sus ventas de lunes a viernes. El programa debe calcular el total semanal, el promedio diario y clasificar el resultado según una meta.

## Entrada / datos
Use esta lista base:

```python
ventas_soda = [85000, 92000, 78000, 110000, 97000]
meta_semanal = 450000
```

## Proceso esperado

- Calcular total semanal.
- Calcular promedio diario.
- Comparar el total con la meta.
- Clasificar:
  - `Meta alcanzada` si el total es mayor o igual a la meta.
  - `Cerca de la meta` si el total alcanza al menos el 80% de la meta.
  - `Requiere atención` si no alcanza el 80%.

## Salida esperada
Debe mostrar total, promedio, porcentaje de cumplimiento y estado final.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|---|---|---|---|
| Caso normal | `[85000, 92000, 78000, 110000, 97000]`, meta `450000` | Meta alcanzada |CORRECT0 |
| Caso límite | total exactamente igual a la meta | Meta alcanzada |CORRECTO|
| Caso especial | total menor al 80% de la meta | Requiere atención |CORRECTO|

````python
# Caso 1 - Escriba su solución aquí

ventas_soda = [85000, 92000, 78000, 110000, 97000]
meta_semanal = 450000

# Complete el análisis
# Caso 1 - Solución para el control de ventas de la soda


#Procesos
total_ventas = sum(ventas_soda)                                                            #LA VARIABLE TOTAL DE VENTAS ES LA SUMA DE LA SEMANA COMPLETA DE MLA SODA                  
promedio_diario = total_ventas / len(ventas_soda)                                          #EL RPOMEDIO ES EL TOTAL DE VENTAS DIVIDIDO ENTRE EL NUMERO DE VENTAS DE LA SODA  FUNCION LEN()
porcentaje_cumplimiento = (total_ventas / meta_semanal) * 100                              #EL PORCENTAJE SE CALCULA DIVIDIENTO EL TOTAL DE VENTAS Y LA META SEMANAL Y ESTE RESULTADO LO MULTIPLICAMOS POR 100

#Clasificación según el estado del negocio                                                # ACA ESTABLESEMOS UNA CONDICIONAL SOBRE LA VARIABLE ESTADO_FINAL 
if total_ventas >= meta_semanal:                                                          # EN DONDE SI LAS VENTAS SEMANALES SUPERAN A LA META SEMANAL NOS INDICA META ALCANSADA 
    estado_final = "Meta alcanzada , su negocio va exelente "                               
elif total_ventas >= (meta_semanal * 0.80):                                               # SI ESO NOS SE CUMPLE PERO LAS EVNTAS SEMANALES SUPERAN EL 80% DE LA META 
    estado_final = "Cerca de la meta , casi lo logramos esforcemonos mas el sigiente mes "# NOS DIRA QUE CASI LO LOGRAMOS Y QUE MAS ESFURZO PARA EL SIGUEINTE MES 
else:
    estado_final = "Requiere atención urgente , algo esta llendo mal "                     # EN ULTIMO CASO SI NO SE CUMPLEN NINGUNA D ELAS ANTERIORES NOS INDICA DEBEMOS ALERTARNOS 
#                                                                                         #ALGO ANDA MAL 
#Salida , resultados esperados 
print("--- REPORTE DE VENTAS SEMANAL ---")
print("----------soda sanpedro---------")
print(f"Total vendido esta semana: ₡{total_ventas}")
print(f"Promedio de ventas al dia: ₡{promedio_diario:.2f}")
print(f"Porcentaje cumplido de la meta: {porcentaje_cumplimiento:.2f}%")
print(f"Estado del negocio: {estado_final}")
print("\n")





#PRUEBA 2

ventas_soda = [85000, 80000, 78000, 110000, 97000]
meta_semanal = 450000


#Procesos
total_ventas = sum(ventas_soda)                                                            #LA VARIABLE TOTAL DE VENTAS ES LA SUMA DE LA SEMANA COMPLETA DE MLA SODA                  
promedio_diario = total_ventas / len(ventas_soda)                                          #EL RPOMEDIO ES EL TOTAL DE VENTAS DIVIDIDO ENTRE EL NUMERO DE VENTAS DE LA SODA  FUNCION LEN()
porcentaje_cumplimiento = (total_ventas / meta_semanal) * 100                              #EL PORCENTAJE SE CALCULA DIVIDIENTO EL TOTAL DE VENTAS Y LA META SEMANAL Y ESTE RESULTADO LO MULTIPLICAMOS POR 100

#Clasificación según el estado del negocio                                                # ACA ESTABLESEMOS UNA CONDICIONAL SOBRE LA VARIABLE ESTADO_FINAL 
if total_ventas >= meta_semanal:                                                          # EN DONDE SI LAS VENTAS SEMANALES SUPERAN A LA META SEMANAL NOS INDICA META ALCANSADA 
    estado_final = "Meta alcanzada , su negocio va exelente "                               
elif total_ventas >= (meta_semanal * 0.80):                                               # SI ESO NOS SE CUMPLE PERO LAS EVNTAS SEMANALES SUPERAN EL 80% DE LA META 
    estado_final = "Cerca de la meta , casi lo logramos esforcemonos mas el sigiente mes "# NOS DIRA QUE CASI LO LOGRAMOS Y QUE MAS ESFURZO PARA EL SIGUEINTE MES 
else:
    estado_final = "Requiere atención urgente , algo esta llendo mal "                     # EN ULTIMO CASO SI NO SE CUMPLEN NINGUNA D ELAS ANTERIORES NOS INDICA DEBEMOS ALERTARNOS 
#                                                                                         #ALGO ANDA MAL 
#Salida , resultados esperados 
print("--- REPORTE DE VENTAS SEMANAL ---")
print("----------soda sanpedro---------")
print(f"Total vendido esta semana: ₡{total_ventas}")
print(f"Promedio de ventas al dia: ₡{promedio_diario:.2f}")
print(f"Porcentaje cumplido de la meta: {porcentaje_cumplimiento:.2f}%")
print(f"Estado del negocio: {estado_final}")
print("\n")



#PRUEBA 3
ventas_soda = [85000, 12000, 78000, 2000, 97000]
meta_semanal = 450000


#Procesos
total_ventas = sum(ventas_soda)                                                            #LA VARIABLE TOTAL DE VENTAS ES LA SUMA DE LA SEMANA COMPLETA DE MLA SODA                  
promedio_diario = total_ventas / len(ventas_soda)                                          #EL RPOMEDIO ES EL TOTAL DE VENTAS DIVIDIDO ENTRE EL NUMERO DE VENTAS DE LA SODA  FUNCION LEN()
porcentaje_cumplimiento = (total_ventas / meta_semanal) * 100                              #EL PORCENTAJE SE CALCULA DIVIDIENTO EL TOTAL DE VENTAS Y LA META SEMANAL Y ESTE RESULTADO LO MULTIPLICAMOS POR 100

#Clasificación según el estado del negocio                                                # ACA ESTABLESEMOS UNA CONDICIONAL SOBRE LA VARIABLE ESTADO_FINAL 
if total_ventas >= meta_semanal:                                                          # EN DONDE SI LAS VENTAS SEMANALES SUPERAN A LA META SEMANAL NOS INDICA META ALCANSADA 
    estado_final = "Meta alcanzada , su negocio va exelente "                               
elif total_ventas >= (meta_semanal * 0.80):                                               # SI ESO NOS SE CUMPLE PERO LAS EVNTAS SEMANALES SUPERAN EL 80% DE LA META 
    estado_final = "Cerca de la meta , casi lo logramos esforcemonos mas el sigiente mes "# NOS DIRA QUE CASI LO LOGRAMOS Y QUE MAS ESFURZO PARA EL SIGUEINTE MES 
else:
    estado_final = "Requiere atención urgente , algo esta llendo mal "                     # EN ULTIMO CASO SI NO SE CUMPLEN NINGUNA D ELAS ANTERIORES NOS INDICA DEBEMOS ALERTARNOS 
#                                                                                         #ALGO ANDA MAL 
#Salida , resultados esperados 
print("--- REPORTE DE VENTAS SEMANAL ---")
print("----------soda sanpedro---------")
print(f"Total vendido esta semana: ₡{total_ventas}")
print(f"Promedio de ventas al dia: ₡{promedio_diario:.2f}")
print(f"Porcentaje cumplido de la meta: {porcentaje_cumplimiento:.2f}%")
print(f"Estado del negocio: {estado_final}")



````
